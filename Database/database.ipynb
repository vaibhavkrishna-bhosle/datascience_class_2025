```python
import pandas as pd
import sqlalchemy
import os

# user = "c##student"
# password = "student123"
# dns = "192.168.1.28/XE"\

user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
dns = os.getenv("DB_DNS")

engine = sqlalchemy.create_engine(f"oracle+oracledb://{user}:{password}@{dns}")

df = pd.read_sql("SELECT * FROM students", engine)
df_short = pd.read_sql("SELECT FULL_NAME, GENDER, DOB, ADMISSION_YEAR, EMAIL FROM students where ADMISSION_YEAR = 2023", engine)
```
